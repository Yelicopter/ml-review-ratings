```python
import re
import nltk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import label_binarize

%matplotlib inline
pd.options.display.max_rows = 10
np.set_printoptions(precision=4, suppress=True)
```

```python
df = pd.read_csv('balanced_reviews.csv')
df
```

## Preprocess the text

```python
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

df["Cleaned_Review"] = df["Review"].apply(clean_text)
df["Rating"] = df["Rating"].astype(int)
```

## Applying TF-IDF

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

tfidf_matrix = tfidf_vectorizer.fit_transform(df["Cleaned_Review"])

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

tfidf_df
```

## Predicting model accuracy

```python
y = df["Rating"]

x_train, x_test, y_train, y_test = train_test_split(tfidf_df, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

## Visualising classifications
### ROC Curve
Compares True Positive Rate (TPR) vs. False Positive Rate (FPR)

```python
# Convert y_test to one-hot encoding (binarized format)
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

# Get probabilities for all classes
y_probs = model.predict_proba(x_test)

# Plot ROC curve for each class
plt.figure(figsize=(8,6))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Rating {i+1} (AUC = {roc_auc:.2f})')

plt.plot([0,1], [0,1], linestyle='--', color='gray')  # Random classifier
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve (OvR)')
plt.legend()
plt.show()
```

This is a Multiclass ROC Curve (One-vs-Rest, OvR). It plots the True Positive Rate (TPR) vs. False Positive Rate (FPR) for each class in a multiclass classification problem.

Each colored curve represents a different class (0-5), showing how well the model distinguishes each class from the others. The AUC (Area Under Curve) values indicate the model's ability to separate each class.

    A higher AUC (closer to 1.0) means better classification for that class.
    A lower AUC (closer to 0.5) means the model struggles to distinguish that class from the others.

### Confusion Matrix
A confusion matrix provides a detailed breakdown of how well your classification model performs by comparing true labels vs. predicted labels.

```python
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.xticks(ticks=np.arange(5), labels=[1, 2, 3, 4, 5])
plt.yticks(ticks=np.arange(5), labels=[1, 2, 3, 4, 5])
plt.title("Confusion Matrix")
plt.show()
```

Most predictions are correct (strong diagonal presence). However, there is some confusion between similar labels.
Example: 

    Class 2 is often misclassified as 1 and 3, meaning the model struggles to differentiate them.
    Similarly, Class 5 is sometimes predicted as 4.

Some classes are more confused than others:

    Class 4 is often mistaken for 5 (877 times).
    Class 1 and 5 are very well classified (2000+ correct predictions, fewer misclassifications).

### Feature Importance
To see which words (features) impact ratings the most.

https://chatgpt.com/share/67c831b3-afb8-8011-b95b-4e98bdeb8067

```python
# Get feature names and coefficients
feature_names = tfidf_vectorizer.get_feature_names_out()
coefficients = model.coef_

importance = np.mean(np.abs(coefficients), axis=0)  # Mean absolute importance

# Get indices of top influential words
top_n = 10
top_indices = np.argsort(importance)[-top_n:]

# Get words and their importance
top_words = [feature_names[i] for i in top_indices]
top_importance = importance[top_indices]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_words, top_importance, color="purple")
plt.xlabel("Absolute Coefficient Value")
plt.ylabel("Words")
plt.title("Top Words Influencing Ratings")
plt.grid(True)
plt.show()
```
